```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
from h5py import File

%matplotlib inline
```

```python
with File("/Users/rstiskalek/Projects/CANDEL/data/CF4/CF4_TFR.hdf5", 'r') as f:
    CF4_TFR = {}
    grp = f["cf4"]
    for key in grp.keys():
        CF4_TFR[key] = grp[key][...]
```

```python
Vcmb = CF4_TFR["Vcmb"]
mag = CF4_TFR['i']
logW = CF4_TFR["lgWmxi"]

m = mag > 8
Vcmb, mag, logW = Vcmb[m], mag[m], logW[m]

```

```python
with plt.style.context("science"):
    plt.figure(figsize=(6, 3.5))
    plt.scatter(Vcmb, mag, s=2, c=logW)

    plt.title("CF4 TFR data")
    plt.colorbar(label=r'$\log\left[W / (\mathrm{km\,s^{-1}})\right]$')
    plt.xlabel(r'$V_{\mathrm{CMB}}\ \left[\mathrm{km\,s^{-1}}\right]$')
    plt.ylabel(r'$m_{\rm obs} \left[\mathrm{mag}\right]$')
    plt.tight_layout()
    plt.savefig("/Users/rstiskalek/Downloads/CF4_TFR_data.png", dpi=500)

    plt.show()
```
